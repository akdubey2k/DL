# Sliding Window Object Detection | Deep Learning Tutorial 30 (Tensorflow, Keras & Python)

__Sliding window object detection__ is a technique that allows you to **_detect objects_** in a picture. This technique is not very efficient as it is very compute intensive. Recently new techniques has been discovered that tried to improve performance such as **R CNN, Fast R CNN, Faster R CNN etc. YOLO** **_(You only look once)_** is a state of the art most modern technique that outperforms all other previous techniques such as **sliding window object detection, R CNN, Fast and Faster R CNN etc.** We will cover YOLO in future videos.
