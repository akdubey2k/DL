# Implementation of activation functions in python

# Sigmoid 

```python
def sigmoid(x):
    return (1/(1 + math.exp(-x)))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(-56)
```

```python
sigmoid(0.5)
```

# tanh 

```python
def tanh(x):
    return ((math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x)))
```

```python
tanh(-56)
```

```python
tanh(50)
```

```python
tanh(1)
```

# ReLU 

```python
def relu(x):
    return (max(0, x))
```

```python
relu(-100)
```

```python
relu(8)
```

# Leaky ReLU 


```python
def leaky_relu(x):
    return (max(0.1 * x, x))
```

```python
leaky_relu(-100)
```

```python
leaky_relu(8)
```
