# MSE without using numpy

```python
import numpy as np

y_pred = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
mse_sum = sum(i**2 for i in (y_true - y_pred))
mse = mse_sum / len(y_true)
```

```python
print('MSE summation :'.ljust(20,'.'), mse_sum)
print('MSE :'.ljust(20, '.'), mse)
```

```python
def mse_wo_numpy(y_true, y_pred):
  mse_sum = 0
  for ytrue, ypred in zip(y_true, y_pred):
    mse_sum += ((ytrue - ypred) ** 2)

  mse = mse_sum/len(y_true)
  print('MSE summation :'.ljust(20,'.'), mse_sum)
  print('MSE : '.ljust(20,'.'), mse)
```

```python
mse_wo_numpy(y_true, y_pred)
```

# MSE by using numpy

## <font color='gold'>**numpy**</font> definition:
* ### numpy.mean(a, axis=None, dtype=None, out=None, keepdims=<no value>, *, where=\<no value>)

Axis or axes along which the means are computed. The default is to compute the mean of the flattened array. If this is a tuple of ints, a mean is performed 
over multiple axes, instead of a single axis or all the axes as before.

```python
# this is not required, because numpy can do the summation and then mean of it by it's own.
mse_np_sum = np.sum(np.square(y_true - y_pred))
mse_np = np.mean(np.square(y_true - y_pred))

# mse_np = np.mean(mse_np_sum) this will give the same summation value, see the definition above ...
```

```python
print('MSE summation :'.ljust(20,'.'), mse_np_sum)
print('MSE : '.ljust(20,'.'), mse_np)
```

```python
print('MSE :'.ljust(20,'.'),np.mean(np.square(y_true - y_pred)))
```
