## Binary Cross Entropy aka Log Loss-The cost function 
### Exercise
#### Implement mean squared error (or MSE) in two ways,
1. Without using numpy (i.e. using plain python)
2. With the use of numpy

## Solution 1: Without using numpy

```python
import numpy as np

y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
mse = sum(2**i for i in (y_true - y_predicted))  
mse/(len(y_true))
```

```python
def mse_wo_numpy(y_true, y_predicted):
    total_error = 0
    for yt, yp in zip(y_true, y_predicted): # two variable to be pass
        total_error += (yt - yp)**2 # operation of two variable mention here
        
    print("Total error:", total_error)
    mse = total_error / len(y_true)
    print("MSE: ", mse)
    return mse
```

```python
mse_wo_numpy(y_true, y_predicted)
```

```python
def mse_with_numpy(y_true, y_predicted): 
    # mse = np.sum((np.array(y_true - y_predicted))**2)/len(y_true)
    mse = np.mean(np.square(y_true - y_predicted))
    return mse
```
