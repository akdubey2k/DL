#### Repeate same price every year, since price per unit is same every year

```python
# price_per_unit = np.array([1000,400,1200],[1000,400,1200],[1000,400,1200]) # array() takes from 1 to 2 positional arguments but 3 were given
# since multiple dimension array and open-close square bracket was not given '[]'
price_per_unit = np.array([[1000,400,1200],[1000,400,1200],[1000,400,1200]])
```

#### Number of units sold every year in multi-dimensional matrix

```python
units_sold = np.array([[30,40,50],[5,10,15],[2,5,7]])
units_sold
```

### Calculate total sales from units and price per unit using matrix multiplication

#### Simple multiplication like this

```python
price_per_unit * units_sold
```

# By using <font color='brown'>Broadcasting,</font> we can avoid using 3 column in <font color='brown'>"price_per_unit"</font> for dot product 
The term __broadcasting__ describes how _**NumPy**_ treats arrays with different shapes during arithmetic operations. Subject to certain constraints, the smaller array is **“broadcast”** across the larger array so that they have compatible shapes.

```python
price_per_unit = ([1000,400,1200])
price_per_unit * units_sold
```

#### In above case numpy is using broadcasting so it expands price_per_unit array from 1 row, 3 columns to 3 row and 3 columns. Correct way to do matrix multiplication is to use dot product as shown below

# By using <font color='brown'>Dot Product,</font> we will get sale of each year

```python
np.dot(price_per_unit, units_sold)
```
