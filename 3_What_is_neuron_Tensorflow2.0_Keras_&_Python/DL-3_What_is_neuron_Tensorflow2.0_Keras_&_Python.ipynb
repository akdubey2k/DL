In this video, we will see how you can think of a __logistic regression as a neuron.__ We will use __insurance dataset__ as a sample and build a __logistic regression. Logistic regression__ uses two step process for classification,

__Step 1: linear regression__ to find best fit line for given dataset<br>
__Step 2: sigmoid or logit function__ to convert this line into values between 0 and 1

Using sigmoid function one can do a classification.
