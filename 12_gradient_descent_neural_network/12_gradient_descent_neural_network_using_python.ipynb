### Now we start implementing gradient descent in plain python. Again the goal is to come up with same w1, w2 and bias that keras model calculated. We want to show how keras/tensorflow would have computed these values internally using gradient descent

### just carried out the <font color='silver'>***weight1, weight2 & bias(intercept)***</font> from the previos program

```python
w1=4.110462
w2=1.2998307
bias=-2.4231539
```

### write routines such as <font color='gold'>sigmoid and log_loss

```python
import math

def sigmoid(z):
    return (1/1+math.exp(-z))

sigmoid(18)
```

### prediction function takes <font color='gold'>age & affordibility

### we will draw the calculation as per this



```python
def pred_fun(age, afford):
    weighted_sum = w1*age + w2*afford + bias
    return sigmoid(weighted_sum)

pred_fun(0.47, 1)
```
