# Pytorch vs Tensorflow vs Keras | Deep Learning Tutorial 6 (Tensorflow Tutorial, Keras & Python)

We will go over what is the difference between __pytorch, tensorflow and keras__ in this video. __Pytorch and Tensorflow__ are two most popular deep learning frameworks. __Pytorch__ is by facebook and __Tensorflow__ is by Google. __Keras__ is not a full fledge deep learning framework, it is just a wrapper around __Tensorflow__ that provides some convenient APIs.
